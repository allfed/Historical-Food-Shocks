```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
os.chdir('../')
print(os.getcwd())
```

```python
# Read the data from results folder
df = pd.read_csv('results/calories.csv', index_col=0)
df.describe()
```

```python
df.head()
```

```python
# Remove the rows with index World, Asia, Americas, Eastern Asia, Southern Asia, Northern America, Europe
df = df.drop(['World', 'Asia', 'Americas', 'Eastern Asia', 'Southern Asia', 'Northern America', 'Europe'])
```

```python
plt.figure(figsize=(10, 60))
sns.heatmap(df)
```

```python
# Sort by the values in 2023 and plot the heatmap again, but only for the top 10
df_sorted = df.sort_values(by='2023', ascending=False)
df_sorted.head(10)
plt.figure(figsize=(10, 8))
sns.heatmap(df_sorted.head(10))
```
