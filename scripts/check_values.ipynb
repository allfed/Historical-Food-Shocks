```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
os.chdir('../')
print(os.getcwd())
```

```python
# Read the data from results folder
df = pd.read_csv('results/calories_by_countries.csv', index_col=0)
df.describe()
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 60))
sns.heatmap(df)
```

```python
# Sort by the values in 2023 and plot the heatmap again, but only for the top 10
df_sorted = df.sort_values(by='2023', ascending=False)
df_sorted.head(10)
plt.figure(figsize=(10, 8))
sns.heatmap(df_sorted.head(10))
```

```python
# Do the same for the yield_changes.csv
df_yield = pd.read_csv('results/yield_changes_by_countries.csv', index_col=0)
df_yield.describe()
```

```python
df_yield.head()
```

```python
df_yield.max().max()
```

```python
df_yield.min().min()
```

```python
plt.figure(figsize=(10, 60))
sns.heatmap(df_yield, vmax=100, vmin=-100, cmap="RdYlBu")
```

```python
# Now only plot values below - 5 
df_yield_below_5 = df_yield[df_yield < -5]
plt.figure(figsize=(10, 60))
sns.heatmap(df_yield_below_5, vmax=0, vmin=-100, cmap="Oranges_r")
```

```python
# Count the yield shocks above 5 for each year and plot as a bar plot
df_yield_below_5_count = df_yield_below_5.count()
plt.figure(figsize=(10, 5))
sns.barplot(x=df_yield_below_5_count.index, y=df_yield_below_5_count.values)
plt.xticks(rotation=90)
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Count of yield shocks below -5%')
```
